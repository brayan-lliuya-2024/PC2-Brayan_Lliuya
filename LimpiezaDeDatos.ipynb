```python
df_cleaned = pd.read_csv("/content/drive/My Drive/Control2/HDI.csv", skiprows=2)
```

```python
df_cleaned = df_cleaned.rename(columns={'Unnamed: 0': 'HDI rank', 'Unnamed: 1': 'Country'})
```

```python
df_cleaned = df_cleaned.iloc[:, :-1]
```

```python
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.iloc[2:]
```

```python
columns_to_drop = ['Unnamed: 3', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 13']
df_cleaned = df_cleaned.drop(columns=columns_to_drop, errors='ignore')
```

```python
df_cleaned.head()
```

```python
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned.info();
```

```python
df_cleaned.dtypes
```

```python
df_cleaned = df_cleaned.drop(columns=['HDI rank'])
```

```python
df_cleaned['Country'] = df_cleaned['Country'].str.strip()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.replace({'—': 0, '..': 0}, inplace=True)
```

```python
df_cleaned['Gross national income (GNI) per capita'] = df_cleaned['Gross national income (GNI) per capita'].str.replace(',', '').astype(float)
```

```python
df_cleaned.head()
```

```python
df_cleaned.dtypes
```

```python
df_cleaned["Life expectancy at birth"] = df_cleaned["Life expectancy at birth"].astype(float)
df_cleaned["Expected years of schooling"] = df_cleaned["Expected years of schooling"].astype(float)
df_cleaned["Mean years of schooling"] = df_cleaned["Mean years of schooling"].astype(float)
df_cleaned["GNI per capita rank minus HDI rank"] = df_cleaned["GNI per capita rank minus HDI rank"].astype(float)
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.columns = df_cleaned.columns.str.strip()
```

```python
df_cleaned["Human Development Index (HDI)"] = df_cleaned["Human Development Index (HDI)"].astype(float)
```
